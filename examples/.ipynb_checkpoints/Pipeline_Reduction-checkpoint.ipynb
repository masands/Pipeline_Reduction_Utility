### Import Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
PATH = "../pipeline_reduction_utility"
os.chdir(PATH)
from pru import pipeline_reduction
```

### Load the Profile Data

```python
PATH = "../data"
df = pd.read_csv(f'{PATH}/profile.csv')
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['X'], df['Y'], label = "Original Pipeline Profile")
plt.legend()
plt.show()
```

### Perform Pipeline Discretisation

```python
pru = pipeline_reduction(df, n_pipe_sec=200)
```

Plot the elevation profile

```python
plt.figure(figsize=(10,5))
plt.plot(df['X'], df['Y'], label = "Original Pipeline Profile")
plt.plot(pru.df_ep['X'], pru.df_ep['Y'], label = "Pipeline Elevation Profile")
plt.legend()
plt.show()
```

Optimise the pipeline profile

```python
pru.solve()
```

Plot the final profile

```python
plt.figure(figsize=(10,5))
plt.plot(df['X'], df['Y'], label = "Original Pipeline Profile")
#plt.plot(pru.df_ep['X'], pru.df_ep['Y'], label = "Pipeline Elevation Profile")
plt.plot(pru.df_result['X'], pru.df_result['Y'], label = "Reduced Pipeline Profile")
plt.legend()
plt.show()
```
